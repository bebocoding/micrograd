```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from micrograd.nn import MLP
from micrograd.engine import Value
```

```python
X, y = make_moons(n_samples=100, noise=0.1, random_state=42)
y = 2*y - 1 # between -1, 1
plt.scatter(X[:,0],X[:,1], c=y)
```

```python
model = MLP(2, [16,16,1])
print("model parameters:", len(model.parameters()))
```

```python
def svm_loss(batch_size=None):
    # batch for stochastic gradient descent
    if batch_size:
        shuffled_idxs = np.random.permutation(X.shape[0])
        Xb, yb = X[shuffled_idxs][:batch_size], y[shuffled_idxs][:batch_size]
    else:
        Xb, yb = X, y
    
    # feed forward loss
    scores = list(map(model, Xb))
    losses = [(1 -yi*scorei).relu() for yi, scorei in zip(yb, scores)]
    data_loss = sum(losses) * (1.0/len(losses))

    # L2 regularization loss
    alpha = 1e-4
    reg_loss = alpha * sum((p*p for p in model.parameters()))
    
    total_loss = data_loss + reg_loss

    # returning accuracy as well
    accuracy = [(yi > 0) == (scorei > 0) for yi, scorei in zip(yb, scores)]
    
    return total_loss, sum(accuracy) / len(accuracy)
```

```python
for k in range(100):
    # forward pass
    loss, accuracy = svm_loss()
        
    # backward pass
    model.zero_grad()
    loss.backward()

    # update
    learning_rate = 1.0 - 0.9*k/100 # learning-rate decay
    for p in model.parameters():
        p.data -= learning_rate * p.grad
        
    print(f"step {k} loss: {loss.data}, accuracy: {accuracy*100}%")
```

```python
# visualize decision boundary

h = 0.25
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Xmesh = np.c_[xx.ravel(), yy.ravel()]
inputs = [list(map(Value, xrow)) for xrow in Xmesh]
scores = list(map(model, inputs))
Z = np.array([s.data > 0 for s in scores])
Z = Z.reshape(xx.shape)

fig = plt.figure()
plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```
